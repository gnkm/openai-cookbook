## 埋め込みを使用したセマンティックテキスト検索

検索クエリを埋め込み、最も類似したレビューを見つけることで、すべてのレビューを非常に効率的かつ低コストでセマンティックに検索できます。データセットは[Get_embeddings_from_dataset Notebook](Get_embeddings_from_dataset.ipynb)で作成されています。

```python
import pandas as pd
import numpy as np
from ast import literal_eval

datafile_path = "data/fine_food_reviews_with_embeddings_1k.csv"

df = pd.read_csv(datafile_path)
df["embedding"] = df.embedding.apply(literal_eval).apply(np.array)
```

ここでは、クエリと文書の埋め込みのコサイン類似度を比較し、上位n件の最適なマッチを表示します。

```python
from utils.embeddings_utils import get_embedding, cosine_similarity

# search through the reviews for a specific product
def search_reviews(df, product_description, n=3, pprint=True):
    product_embedding = get_embedding(
        product_description,
        model="text-embedding-3-small"
    )
    df["similarity"] = df.embedding.apply(lambda x: cosine_similarity(x, product_embedding))

    results = (
        df.sort_values("similarity", ascending=False)
        .head(n)
        .combined.str.replace("Title: ", "")
        .str.replace("; Content:", ": ")
    )
    if pprint:
        for r in results:
            print(r[:200])
            print()
    return results


results = search_reviews(df, "delicious beans", n=3)
```

```python
results = search_reviews(df, "whole wheat pasta", n=3)
```

これらのレビューを簡単に検索することができます。計算を高速化するために、埋め込みベクトルを通じたより高速な検索を目的とした特別なアルゴリズムを使用することができます。

```python
results = search_reviews(df, "bad delivery", n=1)
```

ご覧のとおり、これにより即座に大きな価値を提供できます。この例では、配送失敗の事例を素早く見つけることができることを示しています。

```python
results = search_reviews(df, "spoilt", n=1)
```
