# Sora 2: プロンプトガイド

# 成功する動画プロンプトの作成

## プロンプトを作成する前に
プロンプトの作成は、あなたのストーリーボードを見たことがない撮影監督にブリーフィングするようなものだと考えてください。詳細を省略すると、彼らは即興で対応することになり、あなたが思い描いていたものとは異なる結果になる可能性があります。「ショット」が何を達成すべきかを具体的に説明することで、モデルにより多くのコントロールと一貫性を与えることができます。

しかし、いくつかの詳細をオープンにしておくことも同様に強力です。モデルにより多くの創造的自由を与えることで、驚くような変化や予期しない美しい解釈につながる可能性があります。どちらのアプローチも有効です：**詳細なプロンプトはコントロールと一貫性を提供し、軽いプロンプトは創造的な結果のためのスペースを開きます。** 適切なバランスは、あなたの目標と目指している結果によって決まります。

プロンプトを契約書ではなく、創造的な願望リストとして扱ってください。ChatGPTと同様に、**同じプロンプトを複数回使用すると異なる結果が得られます** – これはバグではなく機能です。各生成は新鮮な解釈であり、時には2番目や3番目の選択肢の方が良い場合もあります。

最も重要なことは、反復する準備をしておくことです。カメラ、照明、アクションの小さな変更でも、結果を劇的に変える可能性があります。モデルと協力してください：あなたが方向性を提供し、モデルが創造的なバリエーションを提供します。

これは正確な科学ではありません—以下のガイダンスは、モデルとの作業から学んだ有用な提案として考えてください。

## APIパラメータ
プロンプトは動画のコンテンツを制御しますが、特定の属性はAPIパラメータによってのみ管理されます。これらは文章で要求することはできず、API呼び出しで明示的に設定する必要があります：
- **model**: `sora-2` または `sora-2-pro`
- **size**: {width}x{height}の形式の文字列。サポートされる解像度はモデルの選択に依存します：
    - sora-2
        - 1280x720, 720x1280
    - sora-2-pro
        - 1280x720, 720x1280
        - 1024x1792, 1792x1024
- **seconds**: クリップの長さ、サポートされる値：「4」、「8」、「12」。デフォルト値は「4」です。

これらのパラメータは動画のコンテナです。解像度、継続時間、品質は「もっと長くして」のような文章によって変更されることはありません。これらはAPI呼び出しで明示的に設定してください。プロンプトは他のすべて（被写体、動き、照明、スタイル）を制御します。

### ビデオ解像度
ビデオ解像度は、Soraにおける視覚的忠実度とモーションの一貫性に直接影響します。高解像度では、詳細、テクスチャ、照明の変化をより正確に生成しますが、低解像度では視覚情報が圧縮され、しばしばぼやけやアーティファクトが生じます。

### 動画の長さ
モデルは一般的に短いクリップでより確実に指示に従います。最良の結果を得るには、簡潔なショットを目指してください。プロジェクトが許可する場合、8秒のクリップを1つ生成するよりも、4秒のクリップを2つ編集でつなぎ合わせる方が良い結果が得られる可能性があります。

## 効果的なプロンプトの構造
明確なプロンプトは、ストーリーボードにスケッチするかのようにショットを描写します。カメラのフレーミングを述べ、被写界深度に言及し、アクションをビートごとに説明し、照明とパレットを設定します。被写体をいくつかの特徴的な詳細で固定することで認識しやすく保ち、単一の説得力のあるアクションによってショットを追いやすくします。

シーケンスをカバーする必要がある場合、単一のプロンプトで複数のショットを記述することも有効です。その際は、各ショットブロックを明確に区別してください：一度に一つのカメラセットアップ、一つの被写体アクション、一つの照明レシピです。これにより、プロジェクトに応じて短いスタンドアロンクリップまたは長い連続的な瞬間を生成する柔軟性が得られます。各ショットを創作単位として扱い、編集で繋ぎ合わせるか、一連のシーケンスとして一気に再生させることができます。

- 短いプロンプトはモデルにより多くの創造的自由を与えます。驚くような結果を期待してください。
- 長く詳細なプロンプトはモデルの創造性を制限します。あなたの指示に従おうとしますが、常に確実に実行できるとは限りません。

短いプロンプトの例を以下に示します：
```text
In a 90s documentary-style interview, an old Swedish man sits in a study and says, "I still remember when I was young."
```
このプロンプトはうまく機能する可能性が高いです：
- `90s documentary`は動画のスタイルを設定します。モデルはカメラレンズ、照明、カラーグレーディングなどの変数をそれに応じて選択します。
- `an old Swedish man sits in a study`は被写体と設定を軽く詳述し、人物と設定がどのように見えるべきかについてモデルに創造的自由を与えます。
- `and says, "I still remember when I was young."`は台詞を記述します。Soraはこれを正確に再現できる可能性が高いです。

このプロンプトは、これらの要件に一致する動画を確実に生成します。しかし、多くの詳細が未定義のため、あなたのビジョンと正確に一致しない可能性があります。例えば、このプロンプトは時間帯、天気、衣装、トーン、キャラクターの外見と年齢、カメラアングル、カット、セットデザイン、その他多くの要因を記述していません。これらの詳細を記述しない限り、Soraがそれらを作り上げることになります。

### 超詳細な指定

複雑で映画的なショットの場合、標準的なプロンプト構造を超えて、ルック、カメラセットアップ、グレーディング、サウンドスケープ、さらにはショットの根拠まで、プロの制作用語で指定することができます。これは、監督がカメラクルーやVFXチームにブリーフィングを行う方法と似ています。レンズ、フィルタリング、照明、グレーディング、モーションに関する詳細な手がかりは、モデルが非常に特定の美学に固定するのに役立ちます。

例えば、**視聴者が最初に気づくもの**、**カメラプラットフォームとレンズ**、**照明の方向**、**カラーパレット**、**テクスチャの質感**、**劇中音**、**ショットのタイミング**を記述することができます。このアプローチは、実際の映画撮影スタイル（例：IMAXの空撮、35mmハンドヘルド、ヴィンテージ16mmドキュメンタリー）にマッチさせたい場合や、ショット間で厳密な連続性を維持したい場合に効果的です。

#### 例

````python
Format & Look
Duration 4s; 180° shutter; digital capture emulating 65 mm photochemical contrast; fine grain; subtle halation on speculars; no gate weave.

Lenses & Filtration
32 mm / 50 mm spherical primes; Black Pro-Mist 1/4; slight CPL rotation to manage glass reflections on train windows.

Grade / Palette
Highlights: clean morning sunlight with amber lift.
Mids: balanced neutrals with slight teal cast in shadows.
Blacks: soft, neutral with mild lift for haze retention.

Lighting & Atmosphere
Natural sunlight from camera left, low angle (07:30 AM).
Bounce: 4×4 ultrabounce silver from trackside.
Negative fill from opposite wall.
Practical: sodium platform lights on dim fade.
Atmos: gentle mist; train exhaust drift through light beam.

Location & Framing
Urban commuter platform, dawn.
Foreground: yellow safety line, coffee cup on bench.
Midground: waiting passengers silhouetted in haze.
Background: arriving train braking to a stop.
Avoid signage or corporate branding.

Wardrobe / Props / Extras
Main subject: mid-30s traveler, navy coat, backpack slung on one shoulder, holding phone loosely at side.
Extras: commuters in muted tones; one cyclist pushing bike.
Props: paper coffee cup, rolling luggage, LED departure board (generic destinations).

Sound
Diegetic only: faint rail screech, train brakes hiss, distant announcement muffled (-20 LUFS), low ambient hum.
Footsteps and paper rustle; no score or added foley.

Optimized Shot List (2 shots / 4 s total)

0.00–2.40 — “Arrival Drift” (32 mm, shoulder-mounted slow dolly left)
Camera slides past platform signage edge; shallow focus reveals traveler mid-frame looking down tracks. Morning light blooms across lens; train headlights flare softly through mist. Purpose: establish setting and tone, hint anticipation.

2.40–4.00 — “Turn and Pause” (50 mm, slow arc in)
Cut to tighter over-shoulder arc as train halts; traveler turns slightly toward camera, catching sunlight rim across cheek and phone screen reflection. Eyes flick up toward something unseen. Purpose: create human focal moment with minimal motion.

Camera Notes (Why It Reads)
Keep eyeline low and close to lens axis for intimacy.
Allow micro flares from train glass as aesthetic texture.
Preserve subtle handheld imperfection for realism.
Do not break silhouette clarity with overexposed flare; retain skin highlight roll-off.

Finishing
Fine-grain overlay with mild chroma noise for realism; restrained halation on practicals; warm-cool LUT for morning split tone.
Mix: prioritize train and ambient detail over footstep transients.
Poster frame: traveler mid-turn, golden rim light, arriving train soft-focus in background haze.
````

## 見た目を導く視覚的な手がかり
プロンプトを書く際、**スタイルは望ましい結果に向けてモデルを導く最も強力な手段の一つ**です。全体的な美的感覚を説明すること - 例えば、*「1970年代の映画」*、*「壮大な、IMAXスケールのシーン」*、または*「16mmモノクロフィルム」* - は、他のすべての選択を枠組みづける視覚的なトーンを設定します。モデルが一貫してそれを維持できるよう、このスタイルを早期に確立しましょう。

洗練されたハリウッドドラマ、手持ちのスマートフォンクリップ、または粒子の粗いヴィンテージコマーシャルのどれを求めるかによって、同じ詳細も全く異なって読み取られます。トーンが設定されたら、ショット、アクション、ライトで具体的な要素を重ねていきます。

明確性が勝利をもたらします。*「美しい街」*のような曖昧な手がかりの代わりに、*「濡れたアスファルト、横断歩道、ネオンサインの反射」*と書きましょう。*「素早く動く」*の代わりに、*「3歩ジョギングして縁石で止まる」*と具体的に指定します。目に見える結果を指し示す動詞と名詞は、常により明確で一貫した出力を提供します。

| **弱いプロンプト** | **強いプロンプト** |
| --- | --- |
| "夜の美しい街" | "濡れたアスファルト、横断歩道、水たまりに映るネオンサイン" |
| "人が素早く動く" | "サイクリストが3回ペダルを漕ぎ、ブレーキをかけて横断歩道で止まる" |
| "映画的な見た目" | "アナモルフィック2.0xレンズ、浅い被写界深度、ボリューメトリックライト" |

カメラの方向とフレーミングは、ショットの感じ方を形作ります。上からのワイドショットは空間と文脈を強調し、目線の高さのクローズアップは感情に注意を集中させます。被写界深度は別の層を追加します：浅いフォーカスは被写体をぼけた背景に対して際立たせることができ、深いフォーカスは前景と背景の両方をシャープに保ちます。ライティングも同様に強くトーンを設定します。柔らかく暖かいキーライトは親しみやすいものを作り出し、冷たいエッジを持つ単一のハードライトはドラマに向かって押し進めます。

キャラクターを導入する際は、ある程度の予測不可能性を期待してください—フレーズの小さな変化が、アイデンティティ、ポーズ、またはシーン自体の焦点を変える可能性があります。ショット間で説明を一貫させ、継続性のためにフレーズを再利用し、競合する可能性のある特徴を混在させることを避けましょう。

**弱い例**

```text
Camera shot: cinematic look
```

**強い例**
```text
Camera shot: wide shot, low angle
Depth of field: shallow (sharp on subject, blurred background)
Lighting + palette: warm backlight with soft rim
```

良いフレーミング指示の例：
- wide establishing shot, eye level 
- wide shot, tracking left to right with the charge
- aerial wide shot, slight downward angle
- medium close-up shot, slight angle from behind

良いカメラモーション指示の例：
- slowly tilting camera
- handheld eng camera

## 動きとタイミングをコントロールする
動きは正しく表現するのが最も困難な部分であることが多いため、シンプルに保ちましょう。各ショットには、明確なカメラの動きを1つと、明確な被写体のアクションを1つ含めるべきです。アクションは、ビートやカウント（小さなステップ、ジェスチャー、ポーズ）で記述すると最も効果的で、時間的に根拠のあるものに感じられます。

「俳優が部屋を横切って歩く」では、作業に必要な情報があまり得られません。「俳優が窓まで4歩歩き、一時停止し、最後の1秒でカーテンを引く」のような記述は、タイミングを正確で実現可能なものにします。

**弱い例**
```text
Actor walks across the room.
```

**強い例**
```text
Actor takes four steps to the window, pauses, and pulls the curtain in the final second.
```

## ライティングと色の一貫性
光は、アクションや設定と同じくらいムードを決定します。フレーム全体に拡散した光は穏やかで中立的な印象を与え、一方で単一の強い光源は鋭いコントラストと緊張感を生み出します。複数のクリップを組み合わせる際、ライティングロジックの一貫性を保つことが、シームレスな編集を可能にします。

光の質と、それを補強する色のアンカーの両方を記述してください。「明るく照らされた部屋」のような大まかな記述ではなく、光源とトーンの組み合わせを具体的に指定します：「暖かいランプの補助光と廊下からの冷たいエッジライトを伴う、柔らかい窓光」のように。3〜5色を指定することで、ショット間でパレットを安定させることができます。

**弱い例**
```text
Lighting + palette: brightly lit room
```

**強い例**
```text
Lighting + palette: soft window light with warm lamp fill, cool rim from hallway 
Palette anchors: amber, cream, walnut brown
```

## 対話と音声
対話は、プロンプト内で直接記述する必要があります。視覚的な描写と話し言葉をモデルが明確に区別できるよう、散文による描写の下に`<dialogue>`ブロック内に配置してください。セリフは簡潔で自然に保ち、タイミングがクリップの長さに合うよう、やり取りを数文に制限するようにしてください。複数キャラクターのシーンでは、話し手を一貫してラベル付けし、交互に発言させてください。これにより、モデルが各セリフを正しいキャラクターのジェスチャーや表情と関連付けやすくなります。

リズムとタイミングについても考慮する必要があります。4秒のショットでは通常1〜2回の短いやり取りに対応でき、8秒のクリップではもう少し多くのやり取りをサポートできます。長く複雑なスピーチは同期がうまくいかず、ペースを崩す可能性があります。

ショットが無音の場合でも、「遠くの交通音のヒス音」や「パリッという音」など、小さな音を一つ提案することでペースを示すことができます。これは完全なサウンドトラックというよりも、リズムの手がかりとして考えてください。

対話を含むプロンプトの例：

```text
A cramped, windowless room with walls the color of old ash. A single bare bulb dangles from the ceiling, its light pooling onto the scarred metal table at the center. Two chairs face each other across it. On one side sits the Detective, trench coat draped across the back of his chair, eyes sharp and unblinking. Across from him, the Suspect slouches, cigarette smoke curling lazily toward the ceiling. The silence presses in, broken only by the faint hum of the overhead light.

Dialogue:
- Detective: "You're lying. I can hear it in your silence."
- Suspect: "Or maybe I'm just tired of talking."
- Detective: "Either way, you'll talk before the night's over."
```

背景音の記述例：

```text
The hum of espresso machines and the murmur of voices form the background.
```

## リミックス機能を使った反復作業
リミックスは微調整のためのもので、ギャンブルではありません。制御された変更を一度に一つずつ行い、何を変更するかを明確に述べましょう：「同じショット、85mmに切り替え」や「同じライティング、新しいパレット：ティール、サンド、ラスト」のように。結果が近づいたら、それを参照として固定し、微調整のみを記述します。そうすることで、すでに機能している要素はすべてロックされたままになります。

ショットが何度も失敗する場合は、シンプルに戻しましょう：カメラを固定し、アクションを簡素化し、背景をクリアにします。一度機能したら、段階的に複雑さを重ねていきます。

# プロンプトテンプレートと例

## プロンプト構造
プロンプトを書く効果的な方法の一つは、モデルに使用してもらいたい異なる種類の情報を分離することです。これは**すべてのケースに適用できる万能の成功レシピではありません**が、明確なフレームワークを提供し、一貫性を保ちやすくします。すべての詳細を含める必要はありません。ショットにとって重要でないものは省略できます。

実際、**特定の要素を曖昧にしておくことで、モデルがより創造的になることを促します**。すべての視覚的選択を厳密に指定しないほど、モデルが解釈し、予期しないが美しい変化で驚かせてくれる余地が生まれます。詳細なプロンプトはより一貫性があり制御された結果をもたらしますが、軽いプロンプトは新鮮で想像力豊かな多様な結果を引き出すことができます。

説明的プロンプトテンプレート：

```text
[平易な言葉でのプロズシーン説明。キャラクター、衣装、風景、天気、その他の詳細を説明する。あなたのビジョンに合うビデオを生成するために、できるだけ詳細に記述する。]

撮影技法：
カメラショット：[フレーミングとアングル、例：ワイド確立ショット、アイレベル]
ムード：[全体的なトーン、例：映画的で緊張感のある、遊び心があり suspenseful、豪華な期待感]

アクション：
```
